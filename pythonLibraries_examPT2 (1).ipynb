```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
orders = pd.read_csv('orders.csv')
items = pd.read_csv('items.csv')
customers =  pd.read_csv('customers.csv')
order_item = pd.read_csv('order_item.csv')
```

```python
#drop duplicates:
import os
items["non_null_count"] = items.notnull().sum(axis=1)
items_sorted = items.sort_values(["non_null_count", "item_name"], ascending=[False, True])
items_new = items_sorted.drop_duplicates(subset=["item_name"], keep="first")
items_new.to_csv("items_new.csv", index=False)
display(items_new)

#id update:
item_mapping = dict(zip(items_new["item_name"], items_new["id"]))
def get_best_item_id(item_id):
    item_names = items.loc[items["id"] == item_id, "item_name"]
    if not item_names.empty: 
        return item_mapping.get(item_names.values[0], item_id)
    return item_id  
order_item["item_id"] = order_item["item_id"].map(get_best_item_id)
order_item.to_csv("order_item_updated.csv", index=False)
```

```python
#delete customers and orders:
customers["nan_count"] = customers.isna().sum(axis=1)
customers_cleaned = customers[customers["nan_count"] < 3]
deleted_customers = set(customers["id"]) - set(customers_cleaned["id"])
orders_cleaned = orders[~orders["customer_id"].isin(deleted_customers)]
deleted_orders = set(orders["order_id"]) - set(orders_cleaned["order_id"])
order_item_cleaned = order_item[~order_item["order_id"].isin(deleted_orders)]
customers_cleaned = customers_cleaned.drop(columns=["nan_count"])
customers_cleaned.to_csv("customers_cleaned.csv", index=False)
orders_cleaned.to_csv("orders_cleaned.csv", index=False)
order_item_cleaned.to_csv("order_item_cleaned.csv", index=False)

display(customers_cleaned)
```
